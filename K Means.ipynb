```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generate simple data
# Create two clusters
data = np.array([
    [1, 2], [1.5, 1.8], [5, 8], [8, 8],
    [1, 0.6], [9, 11], [8, 2], [10, 2],
    [9, 3], [2, 5]
])
```

```python
# Plot the raw data
plt.scatter(data[:, 0], data[:, 1], label='Raw Data', color='blue')
plt.title('Raw Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data)
# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Plot clustered data
for i in range(len(data)):
    plt.scatter(data[i][0], data[i][1], c=('red' if labels[i] == 0 else 'green'))
```

```python
# Plot cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='black', s=200, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
